```python
Steam['Game Type'] = Steam['Cost'].apply(lambda x: 'Paid' if x > 0 else 'Free')
Steam
```

```python
Steam['Steam Revenue'] = Steam.apply(lambda row: row['Cost'] * 0.3 + row['Non-refundable fee'] * row['No. of downloads'], axis=1)
Steam
```

```python
df_yearly_revenue = Steam.groupby('Released year')['Steam Revenue'].sum()
```

```python
df_yearly_revenue
```

```python
Steam['Game Type'].value_counts().plot(kind='bar', color = ("green","red"))
plt.title('Game Type')
plt.show()
```

```python
# Filter data into free and paid games
free_games = Steam[Steam['Cost'] == 0]
paid_games = Steam[Steam['Cost'] > 0]

# Calculate value counts for each dataset
free_game_releases_by_year = free_games['Released year'].value_counts()
paid_game_releases_by_year = paid_games['Released year'].value_counts()

# Create figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for free games
ax[0].pie(
    free_game_releases_by_year, 
    labels=free_game_releases_by_year.index, 
    autopct=lambda p : '{:,.0f} games'.format(p * len(free_games) / 100)
)
ax[0].set_title('Free Game Releases by Year')

# Pie chart for paid games
ax[1].pie(
    paid_game_releases_by_year, 
    labels=paid_game_releases_by_year.index, 
    autopct=lambda p : '{:,.0f} games'.format(p * len(paid_games) / 100)
)
ax[1].set_title('Paid Game Releases by Year')

plt.show()
```

```python
Steam['Genre'].value_counts().plot(kind='barh',color=['orange', 'purple', 'blue'])
plt.title('Genre Distribution')
plt.show()
```

```python
grouped = Steam.groupby('Released year')['Steam Revenue'].mean().reset_index()

# Create line graph
plt.figure(figsize=(10,6))
plt.plot(grouped['Released year'], grouped['Steam Revenue'], marker='o', color = 'red')
plt.title('Revenue Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Revenue (in )')
plt.grid(True)
plt.show()
```

```python
revenue_by_year = Steam.groupby('Released year')['Steam Revenue'].sum()
plt.figure(figsize=(10,6))
revenue_by_year.plot(kind='area',color ='green')
plt.title('Revenue by Year')
plt.xlabel('Released year')
plt.ylabel('Revenue (in )')
plt.show()
```

```python
studio_downloads = Steam.groupby('Studio')['No. of downloads'].sum().reset_index()
```

```python
plt.figure(figsize=(10,8))
plt.pie(studio_downloads['No. of downloads'], labels=studio_downloads['Studio'], 
        autopct='%1.1f%%', pctdistance=0.85, startangle=90)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```
